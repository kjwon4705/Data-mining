- 그래프 (box plot, 평균, 표준편차)

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = sns.load_dataset("iris")
```

```python
data.describe()
```

```python
data['species'].value_counts()
```

```python
versicolor = data.loc[data['species'] == 'versicolor', :]
setosa = data.loc[data['species'] == 'setosa', :]
virginica = data.loc[data['species'] == 'virginica', :]
```

```python
versicolor.describe()
```

```python
setosa.describe()
```

```python
virginica.describe()
```

```python

plt.title("Distribution of Sepal width")
plt.hist(data.sepal_width)
```

```python
sns.scatterplot(x = data.sepal_width, y = data.sepal_length, hue = data.species, s = 60)
```

```python
sns.scatterplot(x = data.petal_width, y = data.petal_length, hue = data.species, s = 60)
```

```python
plt.figure(figsize = (10,10))
sns.pairplot(data, hue = 'species')
```

```python
fig, ax = plt.subplots()

ax.boxplot([data.sepal_length, data.sepal_width, data.petal_length, data.petal_width])
ax.set_ylim(0, 9)
ax.set_xticks([1,2,3,4], ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
ax.set_ylabel('Value (cm)')

plt.show()
```

```python
sns.set(font_scale=1.2)
```

```python
plt.figure(figsize = (9, 3))
plt.subplots_adjust(left=0, bottom=0, right=1, 
                    top=2, wspace=2, hspace=0.6)
plt.subplot(4,1,1)
graph1 = sns.boxplot(y=data['species'], x=data['sepal_length'], palette='Blues')
graph1.set(xlabel = "")
graph1.set_title("Sepal length(cm)")
plt.subplot(4,1,2)
graph2 = sns.boxplot(y=data['species'], x=data['sepal_width'], palette='Blues')
graph2.set(xlabel = "")
graph2.set_title("Sepal width(cm)")
plt.subplot(4,1,3)
graph3 = sns.boxplot(y=data['species'], x=data['petal_length'], palette='Blues')
graph3.set(xlabel = "")
graph3.set_title("Petal length(cm)")
plt.subplot(4,1,4)
graph4 = sns.boxplot(y=data['species'], x=data['petal_width'], palette='Blues')
graph4.set(xlabel = "")
graph4.set_title("Petal width(cm)")
```

```python
plt.figure(figsize = (8,1))
sns.set(font_scale = 1.2)
plt.subplots_adjust(left=0, bottom=0, right=1, 
                    top=4.5, wspace=2, hspace=0.8)
plt.subplot(4,1,1)
graph1 = sns.boxplot(x=data['sepal_length'], palette='Blues')
graph1.set(xlabel = "")
graph1.set_title('Sepal length(cm)')
plt.subplot(4,1,2)
graph2 = sns.boxplot(x=data['sepal_width'], palette='Blues')
graph2.set(xlabel = "")
graph2.set_title('Sepal width(cm)')
plt.subplot(4,1,3)
graph3 = sns.boxplot(x=data['petal_length'], palette='Blues')
graph3.set(xlabel = "")
graph3.set_title('Petal length(cm)')
plt.subplot(4,1,4)
graph4 = sns.boxplot(x=data['petal_width'], palette='Blues')
graph4.set(xlabel = "")
graph4.set_title('Petal width(cm)')
```

```python
plt.figure(figsize = (6,1))
j = sns.boxplot(x=data['petal_width'], palette='Blues')
j.set(xlabel= "")
j.set_title("petal_width")
```
