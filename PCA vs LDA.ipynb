# PCA

```python
from sklearn.preprocessing import StandardScaler 
x = df.drop(['Target'], axis=1).values
y = df['Target'].values 

x = StandardScaler().fit_transform(x) # x객체에 x를 표준화한 데이터를 저장

scaledf = pd.DataFrame(x, columns=list(df.columns[0:13]))
scaledf['target'] = df['Target']
```

```python
scaledf
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
printcipalComponents = pca.fit_transform(x)
principaldf = pd.DataFrame(data = printcipalComponents, columns = ['pca1', 'pca2'])
```

```python
principaldf['Target'] = df['Target']
principaldf
```

```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

import seaborn as sns
#sns.set(font_scale = 1.5)
sns.lmplot(x='pca1', y='pca2', hue='Target', data=principaldf, palette = 'Set2', fit_reg=False)
ax = plt.gca()
ax.set_title('PCA 분석 결과', fontsize = 20)
```

```python
print(pca.explained_variance_ratio_)
```

# LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
# linearComponents = lda.fit(x, df['target'])
out_lda = lda.fit_transform(x, df['target'])
```

```python
lindear_df = pd.DataFrame(data = out_lda, columns = ['lda1', 'lda2'])
linear = pd.concat([lindear_df, df[['Target']]], axis = 1)
linear
```

```python
#sns.set(font_scale = 1.5)
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
sns.lmplot(x='lda1', y='lda2', hue='Target', data=linear, palette = 'Set2', fit_reg=False)
ax = plt.gca()
ax.set_title('LDA 분석 결과', fontsize = 20)
```

```python
print(lda.explained_variance_ratio_)
```
